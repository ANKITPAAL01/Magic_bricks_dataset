```python
##import major libraries

import numpy as np
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv(' MagicBricks.csv')
```

```python
df
```

```python
#Data description and Data information

df.describe()  #it will work with numeric values and continue values
```

```python
#information regarding to dataset

df.info()
```

```python
#chcking missing values in a dataset

df.isnull()
```

```python
df.isnull().sum()
```

```python
#filling null values with appropriate data

df.Bathroom=df.Bathroom.fillna(df.Bathroom.median())
df.Parking=df.Parking.fillna(df.Parking.median())
df.Type=df.Type.fillna('unknown')
df.Per_Sqft=df.Per_Sqft.fillna(df.Per_Sqft.median())
```

```python
#checking null values again

df.isnull().sum()
```

```python
#saving a clean file in local system

df.to_csv('magic_bricks_cleandata.csv')
```
